```python
#IMPORTAMOS LIBRERIAS A UTILIZAR

import pandas as pd
import json
import ast
```

```python
#CARGAMOS EL ARCHIVO JSON Y CONVERTIMOS EL ARCHIVO EN UN DATAFRAME
def l_a_c():
    with open("australian_users_items.json",'rt', encoding='utf-8') as file:
        return[ast.literal_eval(line.strip())
            for line in file]
        
datos = l_a_c()

df_item = pd.DataFrame(datos)

df_item.info()
```

```python
df_item
```

```python
#Fragmentamos el archivo para poder desanidarlo sin problemas con el kernel

chunk_size = 1000  # Elige el tamaño del fragmento

# Dividimos el DataFrame en fragmentos
fragmentos = [df_item[i:i + chunk_size] for i in range(0, len(df_item), chunk_size)]
```

```python
#Creamos la funcion para desanidar

import pandas as pd

def de_nest_items(df):
    # Explotar la columna 'items' en múltiples filas
    items = df['items'].explode()

    # Eliminar la columna 'items' original
    df = df.drop('items', axis=1)

    # Concatenar los datos de 'items' con el DataFrame original
    df = pd.concat([df, items.apply(pd.Series)], axis=1)

    return df
```

```python
fragmentos_desanidados = [de_nest_items(fragmento) for fragmento in fragmentos]
```

```python
df_final = pd.concat(fragmentos_desanidados)
```
